```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\rekha\\Downloads\\SampleSuperstore.csv")
df.head()
```

```python
df.shape
```

```python
df = df.drop(["Country"] , axis = 1)
```

```python
plt.figure(figsize=(20,5))
plt.bar('Sub-Category' , 'Category' , data = df)
plt.show()
```

```python
plt.figure(figsize=(10,10))

df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Category')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x="Segment", hue="Region", data=df)
plt.show()
```

```python
sns.scatterplot("Discount",'Profit',data=df)
```

```python
sns.pairplot(df)
plt.show()
```
